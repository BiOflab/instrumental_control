```python
from pyfirmata import Arduino
import tkinter as tk
import pyfirmata
import time
import numpy as np
import matplotlib.pyplot as plt
```

connect the arduino and choose the analog input

```python
port = 'COM3'    #windows
board = Arduino(port)
pinA0 = board.get_pin('a:0:i')  # a = analogique, 0 = num√©ro de la pin, i = input
```

launch iterator

```python
it = pyfirmata.util.Iterator(board)  
it.start()
```

```python
record and display a temporal tr
```

```python
trace=np.zeros(1000)

for ii in range(1000):

    valeur = pinA0.read()
    trace[ii]=valeur
    time.sleep(0.0001)
    
    
plt.plot(trace)
```
