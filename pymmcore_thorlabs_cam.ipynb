```python
import matplotlib.pyplot as plt
import numpy as np
import pymmcore
import os.path
```

```python
mm_dir = "C:/Program Files/Micro-Manager-2.0"

mmc = pymmcore.CMMCore()
mmc.setDeviceAdapterSearchPaths([mm_dir])
mmc.loadSystemConfiguration(os.path.join(mm_dir, "thorlabs.cfg"))

mmc.snapImage()
mmc.getImage()
```

```python
mmc.setProperty('ThorCam', 'PixelType', '16bit')  # Change pixel type
mmc.setProperty('ThorCam', 'Exposure', '3')  # Change pixel type
```

```python
mmc.snapImage()
ii=mmc.getImage()
plt.imshow(ii)
```
